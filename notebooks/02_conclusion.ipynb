## 核心发现

### 发现一：成长类行业在上涨期弹性最大，但需承担更高回撤风险
计算机、电气设备、电子在上涨期平均日超额收益分别为+0.0047、+0.0033、+0.0029，
胜率达到61%、60%、59%。但在下跌期超额收益转负，说明成长类行业
β值高，涨跌均放大，适合趋势明确的上涨期持有。

### 发现二：银行是最纯粹的防御行业
银行在上涨期胜率仅0.39（27个行业最低），但下跌期胜率0.60（最高），
与市场走势呈现明显负相关。2021-2024累计收益虽然跑输大盘，
但在2022年熊市中相对抗跌，具备典型的防御配置价值。

### 发现三：消费类并非天然防御，受估值影响更大
食品饮料下跌期日超额收益-0.0023，胜率仅0.40，是下跌期表现最差的行业。
与传统认知（消费=防御）相悖，原因在于2021年白酒高估值叠加消费降级，
估值杀跌主导了行情，基本面逻辑让位于资金面逻辑。

### 发现四：周期类行业轮动特征最明显
以化工为代表的周期类在上涨期胜率0.64（最高），
但单日超额收益中位数低于成长类，说明周期股胜在持续性而非弹性。
2021年碳中和行情驱动周期类累计涨幅一度超过40%，
随后随大宗商品周期回落震荡下行。

## 方法论说明

- 超额收益定义：行业日涨跌幅 - 沪深300日涨跌幅
- 数据来源：AkShare，申万一级行业指数，2021年1月至2024年12月
